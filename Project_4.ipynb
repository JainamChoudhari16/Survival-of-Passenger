## ***Project 4: Predicting whether the Passenger has survived or not.***
Here, we have created a ML model predicting whether the passenger has survived or not from Titanic accident.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
df.head()
```

```python
#Checking how may cells have null value in respective columns.
df.isnull().sum()
```

```python
#From the whole data we can say that Survive column is dependent and to predict the survival of a passenger we can use independent columns such as age, gender i.e sex, passenger_class i.e. pclass from the whole dataset.
#We haven't use other columns like name, ticket, fare, etc as they do not help in predicting the survival of the passengers.
```

```python
df.dropna(subset=['Age', 'Sex', 'Pclass'], inplace=True)
x = df[['Age', 'Sex', 'Pclass']] #independent variable
y = df['Survived'] #dependent variable
```

```python
x.isnull().sum()
```

```python
#Replacing null values in age column with mean age values.
x.Age.fillna(x.Age.mean(),inplace = True)
```

```python
x
```

```python
#Converting string value of sex to numeric by using label encoding method
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x.Sex = le.fit_transform(x.Sex) #female = 0 & male = 1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test) #accuracy
```

```python
model.predict([[30, 0, 1]])
```

```python
gender = input("Enter Gender (M/F):").upper()
age = int(input("Enter Age:"))
pclass = int(input("Enter Passenger Class (1/2/3) :"))
if gender == 'M':
  gender = 0
else:
  gender = 1
myprediction = model.predict([[gender,age,pclass]])
if myprediction[0]==1:
  print("The person has Survived")
else:
  print("The person has not Survived")
```
